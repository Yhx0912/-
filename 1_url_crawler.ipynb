```python
## 时间处理函数
def deltime(temp_times):
    times = []
    start_year = 2023
    for i in range(len(list(temp_times))):
        if i == 0:
            times.append(str(start_year)+" "+ temp_times[i]) 
        else:
            if int(temp_times[i][:2]) - int(temp_times[i-1][:2]) > 9:
                start_year = start_year-1
                times.append(str(start_year)+" "+ temp_times[i])
            elif int(temp_times[i][:2]) - int(temp_times[i-1][:2]) < -9:
                start_year = start_year+1
                times.append(str(start_year)+" "+ temp_times[i])
            else:
                times.append(str(start_year)+" "+ temp_times[i])
    return(times)
```
